```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n1 = np.loadtxt('metrics.dat')
n2 = np.loadtxt('metrics2.dat')
n4 = np.loadtxt('metrics4.dat')
n8 = np.loadtxt('metrics8.dat')
n16 = np.loadtxt('metrics16.dat')

metrics = np.concatenate((n1, n2, n4, n8, n16), axis=0).reshape((5, 16, 5))
```

```python
scales = [1, 2, 4, 8, 16]
metric_names = ["Training", "Validation"]
for scale in range(0, 5):
    for i, metric in enumerate([0, 2]):
        plt.plot(metrics[scale, :, metric]*100, label=metric_names[i])
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.title(f"Accuracy {scales[scale]}")
    plt.savefig(f"accuracy_{scales[scale]}.pdf")
    plt.show()
```
